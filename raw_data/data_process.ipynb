```python
name=['202504.xlsx','202503.xlsx','202502.xlsx','202501.xlsx','202412.xlsx','202411.xlsx','202410.xlsx','202409.xlsx','202408.xlsx','202407.xlsx','202406.xlsx','202405.xlsx','202404.xlsx','202403.xlsx','202402.xlsx','202401.xlsx',
'202312.xlsx','202311.xlsx','202310.xlsx','202309.xlsx','202308.xlsx','202307.xlsx','202306.xlsx','202305.xlsx','202304.xlsx','202303.xlsx','202302.xlsx','202301.xlsx',
'202212.xlsx','202211.xlsx','202210.xlsx','202209.xlsx','202208.xlsx','202207.xlsx','202206.xlsx','202205.xlsx','202204.xlsx','202203.xlsx','202202.xlsx','202201.xlsx',
'202112.xlsx','202111.xlsx','202110.xlsx','202109.xlsx','202108.xlsx','202107.xlsx','202106.xlsx','202105.xlsx','202104.xlsx','202103.xlsx','202102.xlsx','202101.xlsx',
]
features=['StockCode', 'StockName', 'MonthlyReturn',
'MonthlyTurnoverRate', 'MonthlyTradingAmount', 'MonthlyTradingVolume',
'ATR', 'ADTM', 'VMA',
'AssetTurnover', 'NetProfit', 'R&D',
'BBIBOLL', 'MACD', 'OperatingProfitRatio',
'TotalRevenue', 'AnnualVolatility', 'Beta24M',
'Beta100W', 'ROA', 'GrossProfit',
'CashFlowPerShare', 'EPS', 'ROC',
'MASS', 'BBI', 'PB',
'PE', 'PEG', 'VR',
'PSY', 'WVAD']
```

```python
# create pairs of names 选取前后两个name
pairs = []
for i in range(len(name)-1):
    pairs.append((name[i], name[i+1]))
pairs
```

```python
import pandas as pd
df =pd.read_excel('202504.xlsx')
# change columns name to features
def process(df):
    df=df.copy()
    df.columns = features
    df=df[:-1]
    # df['rank']等于monthly return的1-10分箱排名
    df['rank10'] = pd.qcut(df['MonthlyReturn'], 10, labels=False)
    df['rank2']= pd.qcut(df['MonthlyReturn'], 2, labels=False)
    return df
process(df)
```

```python
import pandas as pd
for pair in pairs:
    # Read the first Excel file
    df1 = pd.read_excel(pair[0])
    df1=process(df1)
    # Read the second Excel file
    df2 = pd.read_excel(pair[1])
    df2=process(df2)

    # merge rank of df1 to df2 by stock code
    df2 = pd.merge(df2, df1[['StockCode', 'rank10']], on='StockCode', suffixes=('', '_next'))
    df2=pd.merge(df2, df1[['StockCode', 'rank2']], on='StockCode', suffixes=('', '_next'))
    df2= pd.merge(df2, df1[['StockCode', 'MonthlyReturn']], on='StockCode', suffixes=('', '_next'))
    # save df2 to csv
    df2.to_csv(pair[1].replace('.xlsx', '.csv'), index=False)
```
